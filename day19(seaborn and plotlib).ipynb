# **Data visualization with seaborn and plotlib**

```python
"""
objectives:
1.we will visualize various statistical plots using seaborn.
2.we will understand how we can infer insights from visual plots
3.we will visualize how to convert seaborn plots into equivalend plotly plots
"""
import seaborn as sns
sns.__version__
```

```python
'''configure seaborn setup'''
sns.set(
    style="darkgrid",
    palette="flare",
    font="Arial",
    font_scale=1,
)
```

```python
sns.get_dataset_names()
```

```python
""" 
    Statistical plots:
    1.Relational plots(Estimate or visualizes the relationship between 2 or more varibales) 
    2.Distributional plots(visualizes the distribution of one or more variables)
    3.Categorical plots(visualizes the relations between categorical values)

    Features:(x1,x2,x3)=>y
    dataset:height | weight | adress | Healthy(yes/no)
    scenario:is_healthy=model(height,weight)
                       =yes/no
    
"""
tips=sns.load_dataset('tips')
tips.head()
```

```python
fmri=sns.load_dataset('fmri')

fmri.head()
```

```python
penguins=sns.load_dataset('penguins')
penguins.head()
```

# **Relational plot(relplot)**

```python
"""Scatter plot"""
plot=sns.relplot(
    data=tips,#means which dataframe
    x='total_bill',
    y='tip',
    height=5,
    kind='line',
    hue='sex',#coloring based on different categories
    col='day'#creates separate graphs based on different categories
)
plot.set(
    xlabel='Total Bill',
    ylabel='tip',
    title='Total Bill vs Tip'
    
)
```

```python
"""Scatter plot"""
plot=sns.relplot(
    data=tips,#means which dataframe
    x='total_bill',
    y='tip',
    height=3,
    kind='scatter',
    hue='sex',#coloring based on different categories(sex)
    row='day'#creates separate graphs based on different categories(day)
)
plot.set(
    xlabel='Total Bill',
    ylabel='tip',
    title='Total Bill vs Tip'
    
)
```

```python
"""Line plot"""
plot=sns.relplot(
    data=fmri,#means which dataframe
    x='timepoint',
    y='signal',
    height=3,
    kind='line',
    hue='region',#coloring based on different categories
    col='region'#creates separate graphs based on different categories
)
plot.set(
    xlabel='Timeplot',
    ylabel='Signal',
    title='Timeplot vs Signal'
    
)
```

# **Distributional Plot**

```python
""" 
Context: Why we need to understand the distribution of variables?

    A lot of data.
    Some events generates these data. (Data generating process)
    This domain generates the data randomly. So we can not predict what will be the next data.
    Data Science / ML tries to figure our the pattern of the data generation process.
    US (Data Generating Process) : Builds a car.
    Chaina (Machine Learning Model) : Copy the car.
    Car is the data.
    A machine learning tries to mimic the behavior of data generation process.
"""
penguins.head()
```

```python
'''
Distribution plot:
                Univariate distribution: Distribution of one variable/feature/column.
                Bivariate distribution: Distribution of two variables/features/columns.
                Multivariate distribution: Distribution of three or more variables/features/columns.
'''
plot=sns.displot(
    data=penguins,
    x='flipper_length_mm',
    hue='species',
    
)
plot.set(
    xlabel='flipper_length_mm',
    ylabel='count',
)
```

```python
plot=sns.displot(
    data=penguins,
    x='flipper_length_mm',
    hue='species',
    element='step'
)
plot.set(
    xlabel='flipper_length_mm',
    ylabel='count',
)
```

```python
plot=sns.displot(
    data=penguins,
    x='flipper_length_mm',
    hue='species',
    element='step',
    multiple='stack'
)
plot.set(
    xlabel='flipper_length_mm',
    ylabel='count',
)
```

```python
plot=sns.displot(
    data=penguins,
    x='flipper_length_mm',
    hue='species',
    element='step',
    multiple='stack',
    stat='probability',
)
plot.set(
    xlabel='flipper_length_mm',
    ylabel='probability',
    title='hello'
)
```

```python
plot=sns.displot(
    data=penguins,
    x='flipper_length_mm',
    hue='species',
    multiple='stack',
    kind='kde',
    col='species'
)
plot.set(
    xlabel='flipper_length_mm',
    ylabel='count',
)
```

```python
plot=sns.displot(
    data=penguins,
    x='flipper_length_mm',
    hue='species',
    col='species'
)
plot.set(
    xlabel='flipper_length_mm',
    ylabel='count',
)
```

```python
plot=sns.displot(
    data=penguins,
    x='flipper_length_mm',
    hue='species',
    col='species',
    kde=True,
    kind='hist',
    
)
plot.set(
    xlabel='flipper_length_mm',
    ylabel='count',
)
```

```python
plot=sns.displot(
    data=penguins,
    x='species',
    hue='species',
    discrete=True,
    shrink=0.8 ,#(1 is normal)
    
)
plot.set(
    xlabel='species',
    ylabel='count',
)
```

```python
plot=sns.displot(
    data=penguins,
    x='bill_length_mm',
    y='bill_depth_mm',
    cbar=True,#this is called heatmap
)
plot.set(
    xlabel='Bivariate Distribution',
    ylabel='bill_depth_mm',
)
```

```python
sns.pairplot(penguins)
```

```python
g=sns.PairGrid(penguins)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot,fill=True)
g.map_diag(sns.histplot,kde=True)
```

# **Categorical Plot**

```python
""" Strip plot """
plot=sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    kind='strip',
    jitter=False ,#Jitter is the practice of adding a small amount of random noise to data points on a plot. This technique is used to solve the problem of overplotting, which occurs when multiple data points have the same or very similar values and overlap on a char
    hue='tip'
)
plot.set(
    xlabel='day',
    ylabel='Total Bill',
)
```

```python
""" Strip plot """
plot=sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    kind='swarm',
    hue='smoker',
    col='smoker'
)
plot.set(
    xlabel='day',
    ylabel='Total Bill',
)
```

```python
import matplotlib.pyplot as plt
x=tips['day'].value_counts()
y=tips['day'].value_counts().index
print(y)
plt.figure(figsize=(8,8))
plt.pie(
    x=x,
    labels=y,
    autopct='%1.1f%%'
)
plt.title('Distribution of Tips by Day')
plt.show()
```

# **Data Visualization with plotly**

```python
"""Scatter plot"""
plot=sns.relplot(
    data=tips,#means which dataframe
    x='total_bill',
    y='tip',
    height=3,
    kind='scatter',
    hue='sex',#coloring based on different categories
  
)
plot.set(
    xlabel='Total Bill',
    ylabel='tip',
    title='Total Bill vs Tip'
    
)
```

```python
import plotly.express as px
plot=px.scatter(
    data_frame=tips,
    x='total_bill',
    y='tip',
    color='sex',#in plotly hue is called 'color'
)
plot.update_layout(
    xaxis_title='total bill',
    yaxis_title='Tip',
    title='Total bill vs tip'
)
```
