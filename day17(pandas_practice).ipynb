```python
new_data=data.query(
    "Population<1000 and AveBedrms<1"
)
new_data.head()
```

```python
'''Deleting a column'''
new_data1=new_data.drop(
    columns=['MedHouseVal']
)
new_data1.head()
```

```python
new_data1[['Population','AveOccup']].mean()

```

```python
def normalize1(col):
    mean=col.mean()
    std=col.std()
    normalized_col=(col-mean)/std
    return normalized_col 
def normalize2(col):
    min=col.min()
    max=col.max()
    normalized_col=(col-min)/(max-min)
    return normalized_col
    
normalized_data=new_data1[['AveBedrms','Population']].agg(
    [normalize1,normalize2]
)
normalized_data.head()
```
