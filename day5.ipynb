# **Summation of prime numbers**


```python
s=0
print('The prime numbers are:')
for x in range(1,101):
    if x>1:
       is_prime=True 
       for i in range(2,int(x**.5)+1):
         if x%i==0:
           is_prime= False   
       if is_prime:
           s=s+x
           print(x)

                                            
        
print('summantion of all the prime numbers: ',s)
```

# **FACTORIAL OF A NUMBER**

```python
x=int(input('Enter the number'))
s=x

for i in range(x,1,-1):
    s=s*(i-1)
   
if s!=0:
  is_nonzero=True
else:
    is_nonzero=False
if is_nonzero:
    print(f'{x}!={s}')
else:
    print(f'{x}!=1')
```
