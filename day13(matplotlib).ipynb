# **How to use units in y axis**

```python
years=[2014+x for x in range(8)]
income=[55,56,62,61,72,72,73,75]
income_ticks=list(range(40,81,2))
plt.plot(years,income)
plt.title('Income of John(in USD)',fontsize=20,fontname='Times New Roman',fontweight='bold')
plt.xlabel('year')
plt.ylabel('Yearly income(in USD)')
plt.yticks(income_ticks,[f'{x}k USD' for x in income_ticks])
plt.show() 
```

```python
from matplotlib import style
style.use('default')
x1,y1=np.random.random(100),np.random.random(100)
x2,y2=np.arange(100),np.random.random(100)
plt.figure(1)
plt.scatter(x1,y1,s=20)
plt.figure(2)
plt.plot(x2,y2)
plt.show()

```

```python
x1,y1=np.random.random(100),np.random.random(100)
x2,y2=np.arange(100),np.random.random(100)
fig,axis=plt.subplots(1,2,figsize=(15,4))
axis[0].scatter(x1,y1)
axis[0].set_title('Random Scattered',fontname='Times New Roman')
axis[1].plot(x2,y2)
axis[1].set_title('disturbed signal')
plt.show()
```

```python
x1,y1=np.random.random(100),np.random.random(100)
x2,y2=np.arange(100),np.random.random(100)
x3,y3=np.random.random(200),np.random.random(200)
x4,y4=np.arange(300),np.random.random(300)
fig,axis=plt.subplots(2,2,figsize=(15,4))
axis[0,0].scatter(x1,y1)
axis[0,0].set_title('Random Scattered',fontname='Times New Roman')
axis[0,1].plot(x2,y2)
axis[0,1].set_title('disturbed signal')
axis[1,0].scatter(x3,y3)
axis[1,0].set_title('Random Scattered x3',fontname='Times New Roman')
axis[1,1].plot(x4,(y4))
axis[1,1].set_title('disturbed signalx4')
axis[1,1].set_xlabel('test')
axis[1,1].set_ylabel('values')
fig.suptitle('Four plots')
plt.tight_layout()#overlapping dur kore
plt.show()
plt.savefig('fourplots.png')
```

```python
import matplotlib.pyplot as plt
import numpy as np
ax=plt.axes(projection='3d')
x=np.arange(-5,5,.1)#0.1 is the step size
y=np.arange(-5,5,.1)

X,Y=np.meshgrid(x,y)
Z=np.sin(X)*np.cos(Y)
ax.plot_surface(X,Y,Z,cmap='Spectral')
ax.set_title('3D plot')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.view_init(azim=65,elev=20)
plt.show()







```
